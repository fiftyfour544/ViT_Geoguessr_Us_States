```python
import os
import sys
import time
import shutil

os.environ["HF_HUB_ENABLE_HF_TRANSFER"] = "1"

try:
    import hf_transfer
    HF_TRANSFER_AVAILABLE = True
except ImportError:
    HF_TRANSFER_AVAILABLE = False
    os.environ["HF_HUB_ENABLE_HF_TRANSFER"] = "0"

from huggingface_hub import snapshot_download
REPO_ID = "osv5m/osv5m"
LOCAL_DIR = "datasets/osv5m"
MAX_WORKERS = 16

IGNORE_PATTERNS: list[str] = []
MIN_FREE_GB = 280
os.makedirs(LOCAL_DIR, exist_ok=True)

abs_dir = os.path.abspath(LOCAL_DIR)
free_gb = shutil.disk_usage(LOCAL_DIR).free / (1024 ** 3)

if free_gb < MIN_FREE_GB:
    print(f'Free GB: {free_gb:.1f}')

start = time.perf_counter()

download_kwargs = dict(repo_id = REPO_ID, local_dir = LOCAL_DIR, repo_type = "dataset", local_dir_use_symlinks = False, max_workers= MAX_WORKERS)

if IGNORE_PATTERNS:
    download_kwargs["ignore_patterns"] = IGNORE_PATTERNS

try:
    snapshot_download(**download_kwargs)
except KeyboardInterrupt:
    sys.exit(1)
    
elapsed = time.perf_counter() - start
hours, rem = divmod(int(elapsed), 3600)
minutes, secs = divmod(rem, 60)

print(f'\completed in {hours}h {minutes}m {secs}s')
```
